```python
# Import necessary libs
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
# from sklearn.linear_model import LinearRegression
from sklearn import metrics, svm
from sklearn.preprocessing import StandardScaler
# from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import cross_val_score
```

```python
# Read dataframe in
df = pd.read_csv("../data/cloudworker_tasks_adjusted.csv")

# Let's examine the first couple of entries in the dataset
print(f'Dataframe shape - {df.shape}\n')

uniqueUsers = df['user.id'].unique()
print(f'Unique users within set - { uniqueUsers.size }\n')
print(f'Types of data within our set:\n{df.dtypes}\n')
df.head(n=10)
```

```python
# As we can see we have a dataframe with approximately 3883 entries and just over 15 feautre columns within it
# We can see that one of those feature columns is our predictor which is completed_task which is either 0 - did not complete or 1 - did complete
# Let's graph some of the feautres and see what they look like

# First let's get some more information about our data
df.describe()
```

```python
# When trying to predict if a person will complete a task or not, it's probably important to look at the following feautres:
# task.monetary_reward_in_dollars, task.assignable_hits_count, and total.labor.event_count
# Let's visually inspect those features

data_filter_features = df[['task.monetary_reward_in_dollars', 'task.assignable_hits_count', 'total.labor.event_count', 'completed_task']]

sns.pairplot(data_filter_features, hue = 'completed_task', kind = 'scatter', diag_kind='kde');

# Looks like we need some trimming to see a better visualization of our data
```

```python
def trim_outliers(df, column, k=1.5):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    low = Q1 - k * IQR
    upper = Q3 + k * IQR

    # Put the data back into the dataset
    trimmed_df = df[(df[column] >= low) & (df[column] <= upper)]
    return trimmed_df

# Let's trim accordingly
t_df = trim_outliers(df, 'task.monetary_reward_in_dollars', k=2.5)
t_df = trim_outliers(t_df, 'task.assignable_hits_count', k=2.5)
t_df = trim_outliers(t_df, 'total.labor.event_count', k=2.5)

# Let's see the new shape of our dataset
print(f'Dataframe shape - {t_df.shape}\n')

# Look's like we were able to trim out 983 entries that we determined to be outliers
```

```python
data_filter_features = t_df[['task.monetary_reward_in_dollars', 'task.assignable_hits_count', 'total.labor.event_count', 'completed_task']]

# Let's look at the features on the pairplot and see what we can see now in terms of feature vairbles in relation to the completition of the task
sns.pairplot(data_filter_features, hue = 'completed_task');
```

```python
# Support Vector Machine (SVM)
# ----------------------------

# Now that our data is trimmed let's perform a SVM model to categorize
# the 'completed_task'output based on a couple features.

# Let's split our training and test data with our features
# 30% reserved for testing, 70% for training
x_features = ['task.monetary_reward_in_dollars', 'task.assignable_hits_count', 'total.labor.event_count']
y_predictor = 'completed_task'
X_train, X_test, y_train, y_test = train_test_split(t_df[x_features], t_df[y_predictor], test_size=0.3, random_state=42)
```

```python
# Feature scale
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Create and train the SVM model with a linear kernel
# https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html
# linear
svm_clf = svm.SVC(kernel='rbf', C=1.0, random_state=42)
svm_clf.fit(X_train_scaled, y_train)

# Predict
y_pred = svm_clf.predict(X_test_scaled)

# Model Evaluation
print("Test data score of SVM:", svm_clf.score(X_test, y_test))
print("Accuracy of SVM: ", accuracy_score(y_test, y_pred))
print("Precision of SVM: ", metrics.precision_score(y_test, y_pred, average='weighted'))
print("Recall of SVM: ", metrics.recall_score(y_test, y_pred, average='weighted'))
print("F1-Score of SVM: ", metrics.f1_score(y_test, y_pred, average='weighted'))

#linear
# Test data score of SVM: 0.7310344827586207
# Accuracy of SVM:  0.8367816091954023
# Precision of SVM:  0.8387602385762796
# Recall of SVM:  0.8367816091954023
# F1-Score of SVM:  0.8290497749406808

#sigmoid
# Test data score of SVM: 0.2574712643678161
# Accuracy of SVM:  0.7977011494252874
# Precision of SVM:  0.7973612554090772
# Recall of SVM:  0.7977011494252874
# F1-Score of SVM:  0.7975267538644472

#rbf
# Test data score of SVM: 0.6540229885057471
# Accuracy of SVM:  0.8793103448275862
# Precision of SVM:  0.88169672976012
# Recall of SVM:  0.8793103448275862
# F1-Score of SVM:  0.8752647360816025

#poly
# Test data score of SVM: 0.7448275862068966
# Accuracy of SVM:  0.8574712643678161
# Precision of SVM:  0.8667436874634088
# Recall of SVM:  0.8574712643678161
# F1-Score of SVM:  0.8490946311791467

# Todo:
# https://www.geeksforgeeks.org/support-vector-regression-svr-using-linear-and-non-linear-kernels-in-scikit-learn/?ref=lbp
# You can then use metrics like mean squared error (MSE) or mean absolute error (MAE) to measure 
# the error between the predicted and true output values.
```

```python
# Let's get more information about our model
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='BuPu',
            xticklabels=['Positive', 'Negative'],
            yticklabels=['Positive', 'Negative'])

plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
# Ref
# https://www.isanasystems.com/machine-learning-handling-dataset-having-multiple-features/
# https://www.geeksforgeeks.org/cross-validation-machine-learning/

# Validate SVM's robustness with K-Fold Cross Validation technique to prevent overfitting the training
# data with the model to ensure robustness and generalizes well to new data.
scores = cross_val_score(svm_clf, t_df[x_features].values, t_df[y_predictor].values, cv=5)
```

```python
# Print accuracy of each fold (e.g. cv=5 as above, kernal='rbf')
print(scores)

# Print mean and std of all folds.
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))

# 2-folds
# [0.64344828 0.82      ]
# 0.73 accuracy with a standard deviation of 0.09

# 3-folds
# [0.41571872 0.91520165 0.78985507]
# 0.71 accuracy with a standard deviation of 0.21

# 4-folds
# [0.64275862 0.92413793 0.88       0.78758621]
# 0.81 accuracy with a standard deviation of 0.11

# 5-folds
# [0.6        0.86551724 0.89482759 0.8637931  0.79655172]
# 0.80 accuracy with a standard deviation of 0.11
```
